### Problem Statement:

Consider that a real estate company has the data of real-estate prices in Delhi. The company wants to optimise the selling price of the properties, based on important factors such as area, bedrooms, parking, etc.

 

Essentially, the company wants:

    - To identify the variables affecting house prices, e.g., area, number of rooms, bathrooms, etc.
    - To create a linear model that quantitatively relates house prices with variables, such as the number of rooms, area, number of bathrooms, etc.
    - To know the accuracy of the model, i.e. how well do these variables predict the house prices

#### Steps we follow:

    Reading, understanding and visualizing thee data. 
    Preparing the data for modelling (train-test-split)
    Training the data.
    Resedual Analysis
    Predictions and evaluation the test set

# preapring the Data for modelling:

#### Encoding: 
    - Converting binary variables to 1/0
    - 


```python
#yes/no variables
housing.mainroad.value_counts()
```

```python
yes_no_list = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

housing[yes_no_list].head()
```

```python
housing[yes_no_list] = housing[yes_no_list].apply(lambda x: x.map({'yes' : 1, 'no': 0}))
```

```python
housing[yes_no_list].head()
```

```python
housing.head()
```

### Creating Dummies

```python
status = pd.get_dummies(housing["furnishingstatus"])
status.head()
```

```python
# Dropping the redundant variable from dummies
status = pd.get_dummies(housing["furnishingstatus"] , drop_first = True)
status.head()
```

```python
housing = pd.concat([housing, status], axis = 1)
housing.head()
```

```python
housing = housing.drop('furnishingstatus', axis = 1)
housing.head()
```

### Preparing the data for modelling:
    - import the libraries
    - split the data into and train and test sets
    - Rescaling the data 
        * Min-max scaling (Normalization : i.e., bringing the data into 0 and 1) ----> **` preferred `**
        * Standardization (Bringing the data around the mean, so the over mean is 0)

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(housing, train_size = 0.7, random_state = 100)
print(df_train.shape)
print(df_test.shape)
```

### Min-Max scalling:
    - Import the libraries reuired.
    

```python
from sklearn.preprocessing import MinMaxScaler 
```

```python
# Instantiate the object.
scaler = MinMaxScaler()

# create a var-list to fit, we shall only take numeric variables to fit, 
# since the scalling wont show any effect on categorical variables as they are already in between 0 and 1
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking', 'price']

```

```python
df_train[num_vars].head()
```

```python
# fit : learns what needs to be done, i.e., min and maximum values.
# Transform : (x - xmin)/(xmax - xmin)
# fit_transform : Does both .
```

```python
# Lets do both.
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
df_train[num_vars].describe()
```

```python
df_train.head()
```

## Training a model:


```python
plt.figure(figsize = (16,10))
sns.heatmap(df_train.corr(), annot = True, cmap = 'RdYlGn')
plt.show()
```

```python
y_train = df_train.pop("price")
X_train = df_train
```

### Training through statsmodel: Model 1

    - adding one by one 


```python
# import the libraries required

import statsmodels
import statsmodels.api as sm
```

```python
# add a constant
X_train_sm = sm.add_constant(X_train['area'])

# Create the first model
lr = sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.params
```

```python
lr_model.summary()
```

```python
# Add another variable
X_train_sm = X_train[['area', 'bedrooms']]
X_train_sm = sm.add_constant(X_train_sm)

# Create the first model
lr = sm.OLS(y_train, X_train_sm)

# fit the model
lr_model = lr.fit()

lr_model.summary()
```

```python
# add bathrooms
X_train_sm = X_train[['area', 'bedrooms', 'bathrooms']]
X_train_sm = sm.add_constant(X_train_sm)

#create model
lr = sm.OLS(y_train, X_train_sm)

# fit the model
lr_model =  lr.fit()

lr_model.summary()
```

### Training the model through statsmodel: Medel 2
       
    - considering all varaiables and deleting highest collenier one after other


```python
housing.columns
```

```python
# Add all variables at a time
X_train_sm = sm.add_constant(X_train)

#create model
lr = sm.OLS(y_train, X_train_sm)

# fit the model
lr_model =  lr.fit()

lr_model.summary()
```

### Decision to drop based on :
    - p values (drop if the p-value is greater than 0.05 or 50%)
    - VIF (if greater than 10, or even greater than 5)

```python
# import the required libraries 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif["Features"] = X_train.columns
vif["VIF"] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
X_train.shape[1]
```

### Remove 
    1). High p, high VIF
    2). High p-value and low VIF 
    3). High VIF and low p-value



```python
X = X_train.drop('semi-furnished', axis = 1)
```

```python
X_train_sm = sm.add_constant(X)

#create model
lr = sm.OLS(y_train, X_train_sm)

# fit the model
lr_model =  lr.fit()

lr_model.summary()
```

```python
vif = pd.DataFrame()
vif["Features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

```python
#Drop Bedrooms
X = X.drop('bedrooms', axis = 1)
X_train_sm = sm.add_constant(X)

#create model
lr = sm.OLS(y_train, X_train_sm)

# fit the model
lr_model =  lr.fit()

lr_model.summary()
```

```python
vif = pd.DataFrame()
vif["Features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

### Residual Analysis

  

```python
y_train_pred = lr_model.predict(X_train_sm)

res = y_train - y_train_pred

sns.distplot(res)
plt.show()
```

# Model Building through RFE (Recursive feature elimination)

# RFE

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()

lm.fit(X_train, y_train)

rfe = RFE(lm, 10)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building an model using statsmodels for some statistics

```python
X_train_rfe = X_train[col]
X_train_rfe = sm.add_constant(X_train_rfe)
lm = sm.OLS(y_train, X_train_rfe).fit()
lm.summary()
```

```python
X_train_new = X_train_rfe.drop(['bedrooms'], axis = 1)
X_train_lm = sm.add_constant(X_train_new)
lm = sm.OLS(y_train, X_train_lm).fit()
lm.summary()
```

```python
X_train_new.columns
```

```python
X_train_new = X_train_new.drop(['const'], axis = 1)
```

```python
vif = pd.DataFrame()
X = X_train_new
vif["Features"] = X.columns
vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["VIF"] = round(vif["VIF"], 2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```

### Residual Analysis on training set

```python
y_train_pred = lm.predict(X_train_lm)
```

```python
res = y_train - y_train_pred
```

```python
fig = plt.figure()
sns.distplot(res, bins = 20)
fig.suptitle('Error Terms', fontsize = 15)
plt.xlabel('Residuals', fontsize = 15)
plt.show()
```
