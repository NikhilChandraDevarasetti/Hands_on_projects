```python
import pickle
import pandas as pd
```

```python
df = pd.read_csv("attribute_list.csv", header = None)
df
```

```python
df[1][:]
```

```python
ids=list(df[0])
att=list(df[1])

atts=[]
for i,a in zip(ids,att):
    at=[]
    at.append(i)
    at.append(a)
    atts.append(at)

atts
```

```python
atts
```

```python
atts.to_pickle('attribute_list_English.pkl')    #to save the dataframe, df to 123.pkl
df1 = pd.read_pickle('aatribute_list_English.pkl')
df1
```

```python

file = open("attribute_list_English.pkl", 'wb')
```

```python
pickle.dump(atts, file)
```

```python
file.close()
```
